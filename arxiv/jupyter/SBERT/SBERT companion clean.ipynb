```python
hep_tfidf = gensim.models.TfidfModel(hep_bow_corpus, smartirs='npu')
index = gensim.similarities.MatrixSimilarity(hep_tfidf[hep_bow_corpus])
```

```python
tij=hep_abstracts
tij = tij.reset_index(drop=True)
def fastsimplerecommend(string,string1,cutoff):
    pointer3=hep_abstracts.loc[(hep_abstracts['id'].str.match(string))]
    queryvalue3=pointer3['abstract'].values
    processed_query3=preprocess_documents(queryvalue3)
    test_bow_corpus3 = [hep_dictionary.doc2bow(text) for text in processed_query3]
    vec_test_bow3=hep_tfidf[test_bow_corpus3]
    sims = index[vec_test_bow3[0]]
    ind=tij[tij['id'] == string1].index[0]
    if sims[ind]>cutoff:
        return (1,sims[ind])
    return (0,0) 

def fast_sbert_recommender(model,orig,num,cutoff):
    i=1
    j=0
    starter=model.similar_by_word(orig,topn=1000)
    while i<num:
        x=starter[j][0]
        if fastsimplerecommend(orig,x,cutoff)[0]==1:
            #print(f"\n {i}) With TFIDF score {fastsimplerecommend(orig,x,cutoff)[1]} and SBERT rank {j} ")
            print(f"{i}) Paper title : {hep_abstracts.loc[hep_abstracts['id'] == x].title.values}. \n Abstract : {hep_abstracts.loc[hep_abstracts['id'] == x].abstract.values}. \n Authors : {hep_abstracts.loc[hep_abstracts['id'] == x].authors.values}. \n Arxiv ID : {hep_abstracts.loc[hep_abstracts['id'] == x].id.values} \n \n ")
            i=i+1
        j=j+1
```
